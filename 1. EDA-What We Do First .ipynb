#### Missing Values

```python
Missing_col = [feature for feature in df.columns if df[feature].isnull().sum() > 1]
```

```python
for features in Missing_col:
    print(features, np.round(df[features].isnull().median(),4),'% missing')
    
```

#### Find Missing value relationship with diagram

```python
for feature in Missing_col:
    data =df.copy()
    data[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

#### Temporaral Fetaures

```python
year_feature = [feature for feature in numerical_feature if 'Yr' in feature or 'Year' in feature]
year_feature
```

```python
# let's explore the content of these year variables
for feature in year_feature:
    print(feature, df[feature].unique())
```

```python
## Lets analyze the Temporal Datetime Variables
## We will check whether there is a relation between year the house is sold and the sales price

df.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('Year Sold')
plt.ylabel('Median House Price')
plt.title("House Price vs YearSold")
```

```python
## Here we will compare the difference between All years feature with SalePrice

for feature in year_feature:
    if feature!='YrSold':
        data=df.copy()
        ## We will capture the difference between year variable and year the house was sold for
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data['SalePrice'])
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

The numerical features are two types one is discrete another one is continous 

```python
discrete_feature=[feature for feature in numerical_feature if len(df[feature].unique())<25 and feature not in year_feature+['Id']]
print("Discrete Variables Count: {}".format(len(discrete_feature)))
```

```python
df[discrete_feature].head()
```

#### Relationship between discrete features and Sale price

```python
for feature in discrete_feature:
    data = df.copy()
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel('value')
    plt.title(feature)
    plt.show()
```

#### Continous Variable

```python
continuous_feature=[feature for feature in numerical_feature if feature not in discrete_feature+year_feature+['Id']]
print("Continuous feature Count {}".format(len(continuous_feature)))
```

#### Logarthmetic Information

```python
for feature in continuous_feature:
    data = df.copy()
    data[feature]=np.log(data[feature])
    data['SalePrice']=np.log(data['SalePrice'])
    plt.scatter(data[feature],data['SalePrice'])
    plt.xlabel(feature)
    plt.ylabel('Salesprice')
    plt.title(feature)
    plt.show()
```

#### Outliers

```python
for feature in continuous_feature:
    data = df.copy()
    data[feature]=np.log(data[feature])
    data.boxplot(column=feature)
    plt.title(feature)
    plt.ylabel(feature)
    plt.show()
```
